```python
# lets define the map function
def dummies(x,df1):
    temp = pd.get_dummies(df1[x], drop_first = True)
    df1 = pd.concat([df1, temp], axis = 1)
    df1.drop([x], axis = 1, inplace = True)
    return df1
# lets apply the function to the cars_lr

df2 = dummies('Airline_Name',df2)
# df2 = dummies('Trip',df2)
df2 = dummies('Origin',df2)
df2 = dummies('Time_of_Departure',df2)
df2 = dummies('Number_of_Stops',df2)
df2 = dummies('Arrival_Time',df2)
df2 = dummies('Final_Destination',df2)
df2 = dummies('Type_of_Class',df2)
```

#### Train-Test Split

```python
from sklearn.model_selection import train_test_split

np.random.seed(0)
df_train, df_test = train_test_split(df2, train_size = 0.8, test_size = 0.2, random_state = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
num_vars = ['Flight_Duration', 'Numbers_of_day_left','Sale_Price']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
df_train.head()
```

```python
df_train.describe()
```

```python
plt.figure(figsize = (30, 25))
sns.heatmap(df_train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

```python
y_train = df_train.pop('Flight_Duration')
X_train = df_train
```

```python
# lets perform analysis using LinearRegression

from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm 
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
lm = LinearRegression()
lm.fit(X_train,y_train)
rfe = RFE(lm, n_features_to_select=10)
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
X_train.columns[rfe.support_]
```

```python
X_train_rfe = X_train[X_train.columns[rfe.support_]]
X_train_rfe.head()
```

```python
# lets add the constant
# lets fit the model
# lets print the model summary

def build_model(X,y):
    X = sm.add_constant(X) 
    lm = sm.OLS(y,X).fit()
    print(lm.summary())
    return X
    
def checkVIF(X):
    vif = pd.DataFrame()
    vif['Features'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return(vif)
```

```python
X_train_new = build_model(X_train_rfe,y_train)
```

```python
checkVIF(X_train_new)
```

```python
lm = sm.OLS(y_train,X_train_new).fit()
y_train_Flight_Duration = lm.predict(X_train_new)
```

```python
fig = plt.figure()
sns.distplot((y_train - y_train_Flight_Duration), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)
plt.xlabel('Errors', fontsize = 18)
```

Prediction and evaluation

```python
# lets apply Scaling to the test set
num_vars = ['Flight_Duration', 'Numbers_of_day_left', 'Sale_Price']
df_test[num_vars] = scaler.fit_transform(df_test[num_vars])
```

```python
y_test = df_test.pop('Flight_Duration')
X_test = df_test
```

```python
X_train_new = X_train_new.drop('const',axis=1)
# lets create X_test_new dataframe by dropping variables from X_test
X_test_new = X_test[X_train_new.columns]

# lets add a constant variable 
X_test_new = sm.add_constant(X_test_new)
```

```python
y_pred = lm.predict(X_test_new)
```

```python
from sklearn.metrics import r2_score 
r2_score(y_test, y_pred)
```

```python
fig = plt.figure()
plt.scatter(y_test,y_pred)
fig.suptitle('y_test vs y_pred', fontsize=20)
plt.xlabel('y_test', fontsize=18)             
plt.ylabel('y_pred', fontsize=16) 
```
